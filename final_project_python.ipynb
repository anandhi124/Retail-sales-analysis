```python
#Recency
import pandas as pd

# Correct column name (underscore)
df = pd.read_csv("ORDERS.csv")
df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y', errors='coerce')

today = pd.to_datetime('2026-04-05')

# Recency calculation
recency = df.groupby('customer_id').agg({
    'order_date': lambda x: (today - x.dropna().max()).days if not x.dropna().empty else pd.NA
}).reset_index()

# Rename column
recency.columns = ['customer_id', 'recency']

print(recency.head())
```

```python
#Frequency
import pandas as pd

# -----------------------------
# Step 1: Load your dataset
# Replace 'orders.csv' with your file
# -----------------------------
df = pd.read_csv("ORDERS.csv", parse_dates=['order_date'])

# -----------------------------
# Step 2: Compute Frequency for each customer
# -----------------------------
frequency = df.groupby('customer_id').agg({
    'order_id': 'count'  # Number of orders
}).reset_index()

# Rename column
frequency.columns = ['customer_id', 'frequency']

# -----------------------------
# Step 3: Display Frequency
# -----------------------------
print(frequency)
```

```python
import pandas as pd

# Load tables
orders = pd.read_csv("ORDERS.csv")
order_items = pd.read_csv("ORDERS_ITEM.csv")

# Create total price column
order_items['total_price'] = order_items['quantity'] * order_items['list_price']

# Merge orders + order_items
df = pd.merge(orders, order_items, on='order_id')

# Monetary calculation
monetary = df.groupby('customer_id')['total_price'].sum().reset_index()

# Rename column
monetary.rename(columns={'total_price': 'monetary'}, inplace=True)

print(monetary)
```

```python
import pandas as pd
```

```python
orders = pd.DataFrame({
    'customer_id': [1, 2, 1, 3, 2],
    'order_date': ['2026-01-10','2026-01-12','2026-02-01','2026-03-05','2026-03-10'],
    'amount': [100, 200, 150, 300, 50]
})

orders['order_date'] = pd.to_datetime(orders['order_date'])
```

```python
snapshot_date = orders['order_date'].max() + pd.Timedelta(days=1)

rfm_data = orders.groupby('customer_id').agg({
    'order_date': lambda x: (snapshot_date - x.max()).days,
    'customer_id': 'count',
    'amount': 'sum'
}).rename(columns={
    'order_date': 'recency',
    'customer_id': 'frequency',
    'amount': 'monetary'
}).reset_index()

rfm_data['segment'] = 'Gold'
```

```python
rfm_data.head()
```

```python
rfm_data['R_Score'] = pd.qcut(rfm_data['recency'], 3, labels=[3,2,1])
rfm_data['F_Score'] = pd.qcut(rfm_data['frequency'].rank(method='first'), 3, labels=[1,2,3])
rfm_data['M_Score'] = pd.cut(
    rfm_data['monetary'],
    3,
   labels=[1,2,3]
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# RFM Score create
rfm_data['RFM_Score'] = (
    rfm_data['R_Score'].astype(int) +
    rfm_data['F_Score'].astype(int) +
    rfm_data['M_Score'].astype(int)
)

# Segment label function
def segment_label(score):
    if score >= 12:
        return 'High'
    elif score >= 7:
        return 'Medium'
    else:
        return 'Low'

# Apply
rfm_data['segment_label'] = rfm_data['RFM_Score'].apply(segment_label)

# Plot
plt.figure(figsize=(6,4))
sns.countplot(x='segment_label', data=rfm_data, palette='Set2', order=['High','Medium','Low'])
plt.title("Customer Segments")
plt.xlabel("Segment")
plt.ylabel("Count")
plt.show()
```
